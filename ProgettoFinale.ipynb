```python
import pandas as pd
import numpy as np
import os
import plotly.express as px
import plotly.graph_objects as go
import plotly.io as pio
pio.renderers.default='notebook'
df = pd.read_csv("dataset.csv")
```

```python
df['gap'] = df['danceability'] - df['energy']
df_gap = df.sort_values('gap', ascending=False).head(5)
px.bar(df_gap,x=['danceability', 'energy'], y='track_name',barmode='overlay',orientation='h',title='Danceability vs Energy <br>5 canzoni con danceability elevata e con bassa energy contemporaneamente<br>',labels={'value': 'Gap tra danceability e energy', 'variable': 'Categoria', 'track_name': 'Canzone'},)
```

```python
import pandas as pd
import numpy as np
import os
import plotly.express as px
import plotly.graph_objects as go
import plotly.io as pio
pio.renderers.default='notebook'
df = pd.read_csv("dataset.csv")
```

```python
# Calcoliamo la media della popolarità per genere
df_genres = (df.groupby('track_genre')['popularity'].mean().reset_index()).sort_values('popularity').tail(10)
px.bar(df_genres, x='popularity', y='track_genre', range_x=[40,60], labels={'popularity': 'Popolarità media', 'track_genre':'Genere'}, title='Top 10 generi per popolarità media')
```
